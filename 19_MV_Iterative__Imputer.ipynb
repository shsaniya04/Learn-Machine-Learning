# **Missing Value Iterative Imputer**

### **Import Libraries, Load and Analyze Data and Take 5 Sample to apply Iterative Imputer**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/50_Startups_dataset.csv')[['R&D Spend', 'Administration','Marketing Spend', 'Profit']])
np.random.seed(9)
df = df.sample(5)
df
```

### **Select R&D Spend, Administration and Marketing Spend Column**

```python
df = df.iloc[:,0:-1]
df
```

### **Make one - one from each row NaN**

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df.head()
```

### **Fill NaN with Mean Values**

```python
df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
df0
```

```python
df1 = df0.copy()
```

### **Make 1st column Value as NaN**

```python
df1.iloc[1,0] = np.nan
df1
```

```python
X = df1.iloc[[0,2,3,4],1:3]
X
```

```python
y = df.iloc[[0,2,3,4],0]
```

```python
y
```

### **Predict Value**

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df.iloc[1,1:].values.reshape(1, -1))
```
