**Opening a csv file from an URL**

```python
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)
pd.read_csv(data).head(10)
```

**Header parameter**

header = 1: This is the important part for this question. The header parameter tells pd.read_csv() which row in the CSV file should be used as the header (column names) for the DataFrame.

By default, header=0, meaning the first row of the CSV is used as the header.


```python
pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/aug_train.csv', header = 1)
```

**Usecols / Squeeze Parameter**

usecols=['gender']: This argument is key. It tells pd.read_csv to only read the column named 'gender' from the file. By specifying a list of column names here, you can control which data is loaded into your DataFrame, making the process more efficient if you only need a subset of the data.

```python
# pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/aug_train.csv', usecols = ['gender'], squeeze = True)
# Remove the 'squeeze' argument. This will return a DataFrame instead of a Series.
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/aug_train.csv', usecols=['gender'])

# If you need a Series, you can convert the 'gender' column after reading the DataFrame.
gender_series = df['gender']
```

```python
def rename(name):
  if name == "Royal Challengers Bangalore":
    return "RCB"
  else:
    return name

# IN: rename("Royal Challengers Bangalore")
# OUT: 'RCB'

pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/IPL Matches 2008-2020.csv',converters={'team1':rename})
```
