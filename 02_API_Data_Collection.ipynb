```python
import pandas as pd
import requests

# Corrected API key in the URL
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=61ccdc96e27b7165aab3b9994f799f38&language=en-US&page=1')

# Check response status code for errors
if response.status_code == 200:
    df = pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average']]
    print(df.head())
else:
    print(f"Error: {response.status_code} - {response.reason}")
```
