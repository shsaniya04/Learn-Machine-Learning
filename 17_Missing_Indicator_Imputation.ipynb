# **Missing Indicator Imputation**

### **Import libraries, Load and Analyze Dataset**

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/train.csv',usecols=['Age','Fare','Survived'])
df.head()
```

### **Create X & Y for Train-Test-Split**

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

### **Review Without “Missing Indicator Method” Technique**

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

### **Check for Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_trf,y_train)
y_pred = clf.predict(X_test_trf)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### **Missing Indicator**

```python
mi = MissingIndicator()
mi.fit(X_train)
```

```python
MissingIndicator()
```

```python
mi.features_
```

### **Transform: Train Missing**

```python
X_train_missing = mi.transform(X_train)
```

```python
X_train_missing
```

### **Transform: Test Missing**

```python
X_test_missing = mi.transform(X_test)
```

```python
X_test_missing
```

```python
X_train['Age_NA'] = X_train_missing
X_test['Age_NA'] = X_test_missing
```

```python
X_train
```

### **Check Accuracy**

```python
si = SimpleImputer()
X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf  = LogisticRegression()
clf.fit(X_train_trf2,y_train)
y_pred = clf.predict(X_test_trf2)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
