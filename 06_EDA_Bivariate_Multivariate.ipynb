**2. NUMERICAL - CATEGORICAL**

Bar Plot

```python
sns.barplot(x = 'Pclass', y = 'Age', data = titanic, palette = 'Set1')
```

```python
sns.barplot(x = 'Pclass', y = 'Age', data = titanic, hue = 'Sex', palette = 'Set2')
```

```python
sns.barplot(x = 'Pclass', y = 'Fare', data = titanic, hue = 'Sex', palette = 'Set1')
```

Box Plot

```python
sns.boxplot(x = 'Sex', y = 'Age', data = titanic, palette = 'Set1')
```

```python
sns.boxplot(x = 'Sex', y = 'Age', data = titanic, hue = 'Survived', palette = 'Set3')
```

Dist Plot

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

**3. CATEGORICAL - CATEGORICAL**

Cluster Map

```python
titanic.head()
```

```python
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']), cmap = 'magma')
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']), cmap = 'RdBu')
```
