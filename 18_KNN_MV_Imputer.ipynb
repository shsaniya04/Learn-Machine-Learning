# **KNN Missing Value Imputer**

### **Import libraries, Load and Analyze Dataset**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.sample(10)
```

```python
df.isnull().mean()*100
```

### **Create X & Y for Train-Test-Split**

```python
X = df.drop(columns = ['Survived'])
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 2)
```

```python
X_train
```

### **KNN Imputer**

```python
knn = KNNImputer(n_neighbors = 3, weights = 'distance')
X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

### **Convert To DataFrame**

```python
pd.DataFrame(X_train_trf, columns = X_train.columns)
```

### **Logistic Regression**

```python
lr = LogisticRegression()
lr.fit(X_train_trf, y_train)
y_pred = lr.predict(X_test_trf)
accuracy_score(y_test, y_pred)
```
