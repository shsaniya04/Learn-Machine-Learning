Seaborn: Advanced Visualization

Seaborn is built on top of Matplotlib and provides a more aesthetically pleasing and easier way to create statistical plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading Builtin Dataset
df = sns.load_dataset("tips")
print(df.head())
print(df.tail())
print(df.describe())
print(df.info())
```

```python
# Line Plot
sns.lineplot(x = "total_bill", y = "tip", data = df, marker = ".", color = "blue")
plt.title("Line Plot - Total Bill vs Tip")
plt.show()
```

```python
# Scatter plot with hue based on 'sex'
sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = 'sex')
plt.title('Tip vs Total Bill by Gender')
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", data = df, palette = "deep")
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()
```

```python
# Bar Plot
sns.barplot(x = "day", y = "total_bill", data = df, hue = "sex", palette = "viridis")
plt.title("Bar Plot - Average Total Bill by Day")
plt.show()
```

```python
# Histogram
sns.histplot(df["total_bill"], bins = 20, kde = True, color = "blue")
plt.title("Histogram - Total Bill Distribution")
plt.show()
```

```python
# Box Plot 
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = df, palette = "coolwarm")
plt.title("Box Plot - Total Bill Distribution by Day")
plt.show()
```

```python
# Heatmap Example
# For selecting only numeric value we can use this method too
numeric_df = df.select_dtypes(include=["number"])
corr_matrix = numeric_df.corr()

# Drop non-numeric columns
# df_numeric = df.drop(columns=["sex", "smoker", "day", "time"])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()
# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Heatmap - Correlation Matrix")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue = "sex", palette = "husl")
plt.show()
```
