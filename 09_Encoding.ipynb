# **ENCODING**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Learning Machine Learning/customer.csv')
```

```python
df.sample(5)
```

```python
df.info()
```

Eliminating unnecessary coloumns

```python
df = df.iloc[:,2:]
```

```python
df.sample(5)
```

Train-Test-Split

```python
from sklearn.model_selection import train_test_split
```

[:,0:2] selects all rows (:) and columns from index 0 up to (but not including) index 2. This means it's selecting the first two columns as features.

[:,-1:] selects all rows (:) and the last column (-1) as the target variable.

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1:],test_size=0.2)
```

### **Ordinal Encoding**

Import Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
x_train.head()
```

Defining Categories

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
```

Transform Ordinal Encoding

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

```python
x_train
```

```python
oe.categories_
```

### **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
